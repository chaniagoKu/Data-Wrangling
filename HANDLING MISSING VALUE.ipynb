MISSING VALUES
- Tidak memiliki anggota
- Data Kosong/ tidak sama dengan nol

Faktor yang menyebabkan kondisi missing value :
    - Tidak terinput/ tidak terdaftar
    - Data noise (Data string pada RAW)
    - Data corrupted (virus, transfer tidak sempurna, thread, perubahan format data yang tidak sesuai, dll)

Missing values di deskripsikan dengan Null atau NaN
Nilai yang hilang pada data mengakibatkan data tidak dapat dianalisis dengan baik, bisa dianggap sebagai data OUtlier

Tipe-tipe Missing Value :
    - Hilang sepenuhnya secara acak, tidak memiliki titik data hilang dan nilai dalam kumpulan data, hilang atau diamati
    - Hilang secara acak, kecenderungan nilai yang hilang memiliki hubungan sistematis dengand ata yang diamati tetapi tidak dengan data yang hilang
    - Hilang tidak secara acak, ada hubungan yang berbeda antara kecenderungan suatu nilai untuk hilang dan nilainya

## Handling Missing Value

1. Menghapus, jika data bernilai unik dan tidak dimanipulasi (cth : tipe string)
2. Mengisi nilai yang hilang :
    - Nilai konstanta dari kumpulan nilai dari variabel, seperti = 0
    - NIlai rata-rata, median, atau mode untuk kolom
    - Nilai yang diperkirakan oleh model prediktif lain
    - Beberapa imputasi lain

```python
# Menghapus semua elemen data yang hilang (tanpa filter data)
df.dropna()
```

```python
# Data yang semua atributnya hilang menggunakan parameter how ='all'
df.dropna(how ='all')
```

```python
# Menghapus data yang hilang menggunakan ambang batas 'threshold'
df2 = df.dropna(thresh=1)
df2.isnull().sum()
```

```python
df2 = df.dropna(thresh=3)
df2.isnull().sum()
```

```python
df2 = df.dropna(thresh=5)
df2.isnull().sum()
```

```python
df2.dropna(subset = ['density3'])
```

```python
df2.dropna(subset = ['size2'])
df2.isnull().sum()
```

## Mengisi Missing Value

```python
# Mengisi nilai kosong pada kolom tertentu
# Dapat menggunakan nilai ataupun string
df.fillna({'size1':1.0, 'size2':'satu', 'size3':2.0, 'density1':'dua', 'density2':3.0, 'density3':'tiga'})
```

```python
# Menfisi dat ayang hilang dengan memperhatikan urutan pengisiannya
# Bisa diisi dari index data paling akhir ataupun yang lain

# Metode 'ffill' menyebarkan nilai-nilai observasi terakhir ke depan untuk mengisi nilai berikutnya
# Metode 'bfil' 
```

```python
df.fillna('ffill')
```

```python
# Mengisi data yang hilang menggunakan statistika deskriptif (mean, median, modus)
df.fillna(df.mean())
```
