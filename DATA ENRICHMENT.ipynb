### DATA ENRICHING/ENRICHMENT

Tools dan proses yang digunakan untuk meningkatkan atau memperbaiki data mentah (raw data) yang sudah ada
Banyak data yang digabungkan dalam satu tempat
Kumpulan data tersebut menjadi acuan data yang bermanfaat bagi bisnis
Perusahaan menggabungkan informasi internal dan eksternal untuk meningkatkan kualitas keputusan yang diambil berdasarkan data-data tersebut

JENIS-JENIS DATA ENRICHMENT :
    1. Behavioral Enrichment 
        - Penambahan data
        - Data tambahan menjadi informasi tambahan data
        - Dapat memahami karakteristik data
    2. Demographic Enrichment 
        - Ditambah agar penyampaian pesan/informasi pada kelompok demografis tertentu
    3. Geographic Enrichment
        - Penambahan data yang sudah ada, agar bisnis bisa membidik           

        
Bagaimana cara memperkaya ?
Join or Merging
- Merging dua dataset adalah proses menyatukan dua kumpulan data menjadi satu, dan menyelaraskan baris dari masing-masing berdasarkan atribut dan kolom yang sama
- Merge dan Join hal yang serupa

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'key' : ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df2 = pd.DataFrame({'key' : ['a', 'b', 'd'], 'data2': range(3)})
pd.merge(df1,df2)
```

```python
pd.merge(df2,df1)
```

```python
pd.merge(df1,df2, on='key')
```

```python
pd.merge(df1,df2,how='outer')
```

Kebalikan dari relation One To Many dimana setiap baris data dari tabel pertama dihubungkan lebih dari satu baris ke tabel kedua. Hubungan antara file pertama dan file kedua adalah banyak berbanding satu. 
Contoh :
Beberapa orang mengendarai satu mobil.
Beberapa warga melakukan gotong royong di fasum.

Untuk cara menentukan jumlah banyaknya anggota n(A X B),jadi kita perlu melihat banyaknya jumlah anggota A dan B dan kemudian dikalikan (perkalian jumlah anggota himpunan A dan B).

```python
#Many to one
df1 = pd.DataFrame({'key' : ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df2 = pd.DataFrame({'key' : ['a', 'b', 'd'], 'data2': range(3)})
pd.merge(df1,df2,how='cross')
```

MANY TO MANY 

Mempunyai pengertian Satu baris atau lebih data pada tabel pertama bisa dihubugkan ke satu atau lebih baris data pada tabel ke dua. Artinya ada banyak baris di tabel satu dan tabel dua yang saling berhubungan satu sama lain. Hubunga tabel pertama dan tabel kedua adalah banyak berbanding banyak. 
Contoh :
Dalam universitas seorang mahasiswa dapat mengambil banyak mata kuliah 
Satu mahasiswa banyak mengambil banyak mata kuliah dan satu mata kuliah diambil banyak mahasiswa.
