Reshape dengan indeks hirarki :
- Data semi terstruktur ditulis bentuk hirarki, cth : JSON
- Data terstruktur ditulis bentuk tabel
- Pandas + fungsi hirarki ke tabel maupun sebaliknya
- Fungsi yang digunakan :
    - stack

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame(np.arange(9).reshape((3,3)), index=pd.Index(['Ohio','Colorado','Texas'], name = 'state'),
                   columns=pd.Index(['one', 'two', 'three'], name = 'number'))
data
```

```python
#Masuk kedalam bentuk penyimpanan XML
data1 = data.stack()
data1
```

```python
data1.unstack()
```

PIVOT 
- Membentuk kembali data (menghasilkan tabel "poros") berdasarkan kolom
- Memiliki 3 parameter (indeks, kolom, value)
- Value yang digunakan harus unique

```python
data.pivot(columns='one', index='two', values='three')
```

```python
df = pd.read_excel('FILE_LATIHAN/UMKM Kota X.xlsx')
df
```

```python
df.head()
```

```python
df.shape
```

```python
pivot_data = pd.pivot_table(df, index='Sumber Bahan Baku', columns='Jenis Kelamin', aggfunc=np.sum, observed=True)
pivot_data
```

```python
pivot_data = pd.pivot_table(df, index =['Sumber Bahan Baku'], columns=['Jenis Kelamin'], aggfunc={'Jenis Kelamin' : len})
pivot_data
```

```python
pivot_data = pd.pivot_table(df, index =['Sumber Bahan Baku'], columns=['Jenis Kelamin'], aggfunc={'Jenis Kelamin' : len, 'Usia' : np.mean})
pivot_data
```

```python
pivot_data = pd.pivot_table(df, index =['Lokasi UMKM'], aggfunc={'Lokasi UMKM' : len, 'Usia' : np.mean})
pivot_data
```
